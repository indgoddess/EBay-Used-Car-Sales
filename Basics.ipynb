### Split the odometer_km into groups, and use aggregation to see if average prices follows any patterns based on the mileage.
    

```python
auto_copy["odometer_km"].value_counts()
```

```python
ac = auto_copy
bins = [0, 30000, 60000, 80000, 100000,125000, 150000]
names = ['<30k', '30k-60k', '60k-80k', '80k-100k', '100k-125k','125k-150k']

ac['odometer_km_range'] = pd.cut(ac['odometer_km'], bins, labels=names)
```

```python
ac['odometer_km_range'].value_counts()
```

```python
ac[(ac['odometer_km'] >= 30000) & (ac['odometer_km']  < 60000)].mean()
```

```python
ac[(ac['odometer_km'] >= 125000) & (ac['odometer_km']  < 150000)].mean()
```

```python
ac[(ac['odometer_km'] >= 80000) & (ac['odometer_km']  < 100000)].mean()
```

```python
ac[(ac['odometer_km'] >= 60000) & (ac['odometer_km']  < 80000)].mean()
```

#### Observation - We can derive that price is inversely proportional to odometer ratings.
